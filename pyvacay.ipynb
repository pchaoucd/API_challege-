```python
hotel_df.index=range(len(hotel_df))
hotel_df
hotel_df
```

```python
#Set parameters to search for hotels with 5000.
target_radius = 5000
target_type = 'lodging'

url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

for hotel in hotel_df.index:
    target_coordinates = hotel_df['Lat'][hotel].astype(str) + "," + hotel_df['Lng'][hotel].astype(str)



    params = {
      "location": target_coordinates,
      "radius": target_radius,
      "type": target_type,
      "key": g_key
    }


    response = requests.get(url, params=params)
    locations = response.json()

hotel_df['Hotel Name'][hotel] = locations["results"][0]["name"]
```
